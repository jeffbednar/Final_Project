```python
import pandas as pd
import numpy as np
```

```python
vacinations_df = pd.read_csv("country_vaccinations_raw.csv")
cases_deaths_df = pd.read_csv("worldometer_coronavirus_daily_data_raw.csv")
```

```python
vacinations_df
```

```python
cleaned_vacinations_df = vacinations_df.drop(["iso_code", "source_name", "source_website", "daily_vaccinations_per_million", "people_fully_vaccinated_per_hundred", "people_vaccinated_per_hundred", "total_vaccinations_per_hundred"], axis=1)
```

```python
cleaned_vacinations_df
```

```python
cases_deaths_df
```

```python
cleaned_cases_deaths_df = cases_deaths_df[cases_deaths_df["date"] > "2021-1-01"]
```

```python
cleaned_cases_deaths_df
```

```python
combined_df = pd.merge(cleaned_vacinations_df, cleaned_cases_deaths_df, how='inner', on="country")
```
